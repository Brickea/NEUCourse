## Join


* Write a query in SQL to find 
    * the name of all reviewers 
    * who have rated 7 or more stars to their rating.

```python
%%sql
SELECT DISTINCT * FROM reviewer;
```

```python
%%sql
SELECT * FROM reviewer rv
JOIN rating ra
ON ra.rev_id = rv.rev_id AND ra.rev_stars>= 7;
```

* Write a query in SQL to list the first and last names of all the actors who were cast in the movie 'Annie Hall', and the roles they played in that production.

```python
%%sql
SELECT m.mov_title, temp.act_fname, temp.act_lname, temp.role FROM
(SELECT a.act_id, m.mov_id, a.act_fname, a.act_lname, m.role FROM movie_cast m
JOIN actor a
ON m.act_id = a.act_id) temp
JOIN movie m
ON temp.mov_id = m.mov_id AND m.mov_title='Annie Hall';
```

* Write a query in SQL to list all the actors who have not acted in any movie between 1990 and 2000


```python
%%sql
SELECT act_fname, act_lname FROM
(SELECT mc.act_id, mc.mov_id FROM movie m
JOIN movie_cast mc
ON m.mov_id = mc.mov_id
WHERE mov_year>=1990 AND mov_year <=2000) temp
RIGHT JOIN actor a
ON temp.act_id = a.act_id
WHERE mov_id IS NULL
```

* Write a query in SQL to find all the years which produced a movie that received a rating of 3 or 4, and sort the result in increasing order.


```python
%%sql
SELECT m.mov_title, r.rev_stars FROM rating r
JOIN movie m
ON r.mov_id = m.mov_id
WHERE r.rev_stars >=3 AND r.rev_stars <= 4
```

* Write a query in SQL to find the first and last name of an actor with their role in the movie which was also directed by themselves


```python
%%sql
SELECT a.act_id,mc.mov_id, a.act_fname, a.act_lname, mc.role FROM movie_cast mc
JOIN actor a
ON mc.act_id = a.act_id
```

```python
%%sql
SELECT md.dir_id, temp.act_id, temp.act_fname, temp.act_lname,temp.role FROM movie_direction md
JOIN (SELECT a.act_id,mc.mov_id, a.act_fname, a.act_lname, mc.role FROM movie_cast mc
JOIN actor a
ON mc.act_id = a.act_id) temp
ON temp.mov_id = md.mov_id
```

```python
%%sql
SELECT * FROM director
```

```python
%%sql
SELECT d.dir_fname director_actor_fname, d.dir_lname director_actor_lname, temp1.role FROM
(SELECT md.dir_id, temp.act_id, temp.act_fname, temp.act_lname,temp.role FROM movie_direction md
JOIN (SELECT a.act_id,mc.mov_id, a.act_fname, a.act_lname, mc.role FROM movie_cast mc
JOIN actor a
ON mc.act_id = a.act_id) temp
ON temp.mov_id = md.mov_id) temp1
JOIN director d
WHERE d.dir_fname = temp1.act_fname AND d.dir_lname = temp1.act_lname
```

* Write a query in SQL to generate a report which contain the columns 
    * movie title, 
    * name of the female actor, 
    * year of the movie, 
    * role, 
    * movie genres, 
    * the director, 
    * date of release,
    * rating of that movie

```python
%%sql
SHOW tables;
```

```python
%%sql
SELECT mov_title,act_fname,act_lname,mov_year,role,gen_title,dir_fname,dir_lname,mov_dt_rel,rev_stars FROM movie
NATURAL JOIN movie_cast
NATURAL JOIN movie_genres
NATURAL JOIN movie_direction
NATURAL JOIN rating
NATURAL JOIN director
NATURAL JOIN genres
NATURAL JOIN actor
WHERE act_gender = "F"
```

* Write a query in SQL to compute a report which contain the genres of those movies with their average time and number of movies for each genres.

```python
%%sql
SELECT gen_title, AVG(mov_time) average_time, COUNT(*) number FROM movie
NATURAL JOIN movie_genres
NATURAL JOIN genres
GROUP BY gen_title
```

* Write a query in SQL to return the name of all reviewers and name of movies together in a single list.

```python
%%sql
SELECT rev_name, mov_title FROM movie
NATURAL JOIN rating
NATURAL JOIN reviewer
WHERE rev_name IS NOT NULL
```
