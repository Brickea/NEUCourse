# 1. Number of prime numbers

In number theory, the prime number theorem (PNT) describes the asymptotic distribution of  prime numbers among the positive integers. It formalizes the intuitive idea that primes become less common as they become larger by precisely quantifying the rate at which this occurs. The theorem was proved independently by [Jacques Hadamard](https://en.wikipedia.org/wiki/Jacques_Hadamard) and [Charles Jean de la Vallée Poussin](https://en.wikipedia.org/wiki/Charles_Jean_de_la_Vall%C3%A9e_Poussin) in 1896 using ideas introduced by [Bernhard Riemann](https://en.wikipedia.org/wiki/Bernhard_Riemann) (in particular, the [Riemann zeta function](https://en.wikipedia.org/wiki/Riemann_zeta_function)).

Roughly speaking, around a large x, the probability that an integer is a prime is 1/log(x). 

Given a `range(1,30)`, write a list comprehension that retuns all numbers in that range that are divisible by 8 (whose remainder by division by 8 is zero) 

````python
ints = range(1, 30)
divisiable_8 = [i for i in ints if i%8==0]
divisiable_8
````

Use the code above (but replace the 8 appropriately) to produce all prime numbers in that range. Use a single line of python.

````python
list(filter(lambda x:all(map(lambda p:x % p != 0,range(2,x))),ints))
````

Now use the code above to find out the number of primes between 1 and 10,000.

````python
ints = range(1, 10000)
primes_number = list(filter(lambda x:all(map(lambda p:x % p != 0,range(2,x))),ints))
primes_number
````

Verify that the prime number theorem is correct.

````python
import math
print("the probability that an integer is a prime based on 1/log(x).")
print(1/math.log(10000))
print("the probability that calculate is")
print(len(primes_number)/len(ints))
````

What's the diagonal matrix made out of the inverse of the number of interactions per character in the diagonal?

````python
Adj.todense()
pd.DataFrame(Adj.todense())
np.seterr(divide='ignore')  # ignore division-by-zero errors

degrees = np.ravel(Adj.sum(axis=1))
Deginv = sparse.diags(1 / degrees).tocsr()
````

Use the function below to plot it.

````python
import matplotlib.pyplot as plt
from scipy.sparse import coo_matrix

def plot_coo_matrix(m):
    if not isinstance(m, coo_matrix):
        m = coo_matrix(m)
    fig = plt.figure()
    fig, ax = plt.subplots(figsize=(10, 5))
    # you may have to replace 'axisbg' below with 'facecolor', depending on the version of your matplotlib
    ax = fig.add_subplot(111, facecolor='black') #axisbg='b')
    ax.plot(m.col, m.row, 's', color='white', ms=1)
    ax.set_xlim(0, m.shape[1])
    ax.set_ylim(0, m.shape[0])
    ax.set_aspect('equal')
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    return ax
````

Now plot the adjacency matrix.

````python
A = Adj.tocoo()
ax = plot_coo_matrix(A)
ax.figure.show()
````

````python
def pagerank_plot(in_degrees, pageranks, names, *,
                  annotations=[], **figkwargs):
    """Plot node pagerank against in-degree, with hand-picked node names."""

    fig, ax = plt.subplots(**figkwargs)
    ax.scatter(in_degrees, pageranks, c=[0.835, 0.369, 0], lw=0)
    for name, indeg, pr in zip(names, in_degrees, pageranks):
        if name in annotations:
            text = ax.text(indeg + 0.1, pr, name)

    ax.set_ylim(0, np.max(pageranks) * 1.1)
    ax.set_xlim(-1, np.max(in_degrees) * 1.1)
    ax.set_ylabel('PageRank')
    ax.set_xlabel('In-degree (number of interactions)')
````
