## 1. Convert picture to ASCII

Convert color pictures to grayscale

* From RGB in color picture to gray value gray conversion formula is:

> Gray = 0.299R+0.587G+0.114B

```python
char_string = 'asdfzcv?-_+~<>i;:,"^'
img_path='./data/test.jpg'
save_path = './'
def RGB_to_char(r,g,b):
    length = len(char_string)
    gray = int(0.2126 * r + 0.7152 * g + 0.0722 * b)

    # The 
    unit = (256.0 + 1) / length

    # gray值对应到char_string中的位置（索引值）
    idx = int(gray / unit)
    return char_string[idx]
```

RGB value based on color depth (range from 0 to 255, where 0 is black and 255 is white)

```python
from PIL import Image

#预处理（将图片尺寸压缩，并转为灰度图） 
def preprocess(img_path,delta=100):
    img = Image.open(img_path) 
    # 获取图片尺寸
    width, height = img.size
    # 获取图片最大边的长度 
    if width > height:
        max = width
    else:
        max = height

    # 伸缩倍数scale
    scale = max / delta
    width, height = int(width / scale), int(height / scale)
    img = img.resize((width, height)) 
    return img
```

Involving your favorite character set

According to the sequence and length of character set, RGB value is encoded into corresponding characters.

```python
def img2char(img_obj, savepath):
    txt = ''
    width, height = img_obj.size
    # 获取像素点的rgb元组值，如(254, 0, 0)，并将其转化为字符
    for i in range(height):
        line = ''
        for j in range(width):
            line += RGB_to_char(*img_obj.getpixel((j, i)))
        txt = txt + line + '\n'
    print(txt, end='')

#     # 保存字符画
#     with open(savepath, 'w+', encoding='utf-8') as f:
#         f.write(txt)

img_obj = preprocess(img_path)
img2char(img_obj, save_path) 
```

```python
import matplotlib.pyplot as plt
from scipy.sparse import coo_matrix

def plot_coo_matrix(m):
    if not isinstance(m, coo_matrix):
        m = coo_matrix(m)
    fig = plt.figure()
    fig, ax = plt.subplots(figsize=(10, 5))
    # you may have to replace 'axisbg' below with 'facecolor', depending on the version of your matplotlib
    ax = fig.add_subplot(111, facecolor='black') #axisbg='b')
    ax.plot(m.col, m.row, 's', color='white', ms=1)
    ax.set_xlim(0, m.shape[1])
    ax.set_ylim(0, m.shape[0])
    ax.set_aspect('equal')
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    return ax
```

```python
def pagerank_plot(in_degrees, pageranks, names, *,
                  annotations=[], **figkwargs):
    """Plot node pagerank against in-degree, with hand-picked node names."""

    fig, ax = plt.subplots(**figkwargs)
    ax.scatter(in_degrees, pageranks, c=[0.835, 0.369, 0], lw=0)
    for name, indeg, pr in zip(names, in_degrees, pageranks):
        if name in annotations:
            text = ax.text(indeg + 0.1, pr, name)

    ax.set_ylim(0, np.max(pageranks) * 1.1)
    ax.set_xlim(-1, np.max(in_degrees) * 1.1)
    ax.set_ylabel('PageRank')
    ax.set_xlabel('In-degree (number of predators)')
```

```python
def power(M, damping=0.85, max_iter=10**5):
    n = M.shape[0]
    r0 = np.full(n, 1/n)
    r = r0
    for _iter_num in range(max_iter):
        rnext = damping * M @ r + (1 - damping) / n
        if np.allclose(rnext, r):
            break
        r = rnext
    return r
```

```python
def power2(M, damping=0.85, max_iter=10**5):
    n = M.shape[0]
    dangling = (1/n) * np.ravel(M.sum(axis=0) == 0)
    r0 = np.full(n, 1/n)
    r = r0
    for _ in range(max_iter):
        rnext = (damping * (M @ r + dangling @ r) +
                 (1 - damping) / n)
        if np.allclose(rnext, r):
            return rnext
        else:
            r = rnext
    return r
```

```python
pagerank_power = power(Trans)
pagerank_power2 = power2(Trans)
np.corrcoef([pagerank, pagerank_power, pagerank_power2])
```
