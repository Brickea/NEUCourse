## Build the Bayesian model

As we know, a count outcome is most often modelled as a Poisson distribution. Modeling team strengths is usually accomplished with a vector of goals scored $y$ as independent Poisson distributions: $(y\;|\;θ) \propto Poisson(θ)$ 

I will model $\theta$ as an exponential of two independent variables: an tracks variable **track**, and a horses variable **horses**, times a constant factor k:

$$θ = k * e^{\text{track}_g} * e^{\text{horse}_g}$$

$$\log(θ) = \text{track}_g + \text{horse}_g$$

The resulting intensity is determined jointly by an track and a horse type for each horse, represented by parameters `track` and `horse`, respectively

$track_t \propto \text{Normal}(μ_{track},τ_{track})$ and $horse_t \propto \text{Normal}(μ_{horse}, τ_{horse})$.

```python
observed_result = df_need['Result'].values

trackNumber = df_need['trackNumber'].values
horseNumber = df_need['horseNumber'].values

num_horse = len(horses)
num_track = len(tracks)

num_match = len(trackNumber)
num_horse, num_match
```

```python
observed_result.sum()
```

```python
with pm.Model() as model:

    sd_track = pm.HalfStudentT('sd_track', nu=3, sd=2.5)
    sd_horse = pm.HalfStudentT('sd_horse', nu=3, sd=2.5)
    intercept = pm.Flat('intercept') #no idea

    track_star = pm.Normal("track_star", mu=0, sd=sd_track, shape=num_track)
    horse_star = pm.Normal("horse_star", mu=0, sd=sd_horse, shape=num_horse)

    trackss = pm.Deterministic('atts', track_star - tt.mean(track_star))
    horsess = pm.Deterministic('defs', horse_star - tt.mean(horse_star))
    
    result_theta = tt.exp(intercept + trackss[trackNumber] + horsess[horseNumber])

    result_points = pm.Poisson('home_points', mu=result_theta, observed=observed_result)
```

```python
with model:
    trace = pm.sample(1000, tune=1000, cores=2)
    pm.traceplot(trace)
```
