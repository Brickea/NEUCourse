## New England Patriots

Here's their schedule for this season:

````python
from sportsreference.nfl.schedule import Schedule

ne_schedule = Schedule('NWE')
for game in ne_schedule:
    print(game.date, game.points_scored, game.points_allowed)
````

Here's the schedule for ***all*** NFL teams:

````python
from sportsreference.nfl.teams import Teams

for team in Teams():
    print(team.name)
    schedule = team.schedule  # Request the current team's schedule
    for game in schedule:
        print(game.date, game.points_scored, game.points_allowed)
````

## Players

Teamwork is paramount to success in sports, but a team can’t exist without individual contributions. And maybe player stats can help predict game results. Maybe.

You can dive into a player’s career and identify record performances, career totals, and yearly trends from sportsreference's `Player` module. 

To get started, you must first find a player’s ID. In general, the player ID is a string in the format Llll(l)Ffnn where Llll(l) is the first four or five letters of the player’s last name, Ff is the first two letters of the player’s first name, and nn is a number starting at 00 for the first time that player ID has been used and increments by 1 for every successive player. Camel-cased.

So for example, Tom Brady might be `BradTo00`. You can also navigate to [sportsreference.com](sportsreference.com,), find the patriots, and click on Tom Brady to find his ID fro the [URL](https://www.pro-football-reference.com/players/B/BradTo00.htm).

For a list of all APIs on Tom Brady, use Python's `dir` module:

````python
from sportsreference.nfl.roster import Player

tombrady = Player('BradTo00')
dir(tombrady)
````

Here are the stats for Tom Brady:

````python
from sportsreference.nfl.roster import Player

tombrady = Player('BradTo00')
print(tombrady.name)  # Prints 'Drew Brees'
print(tombrady.passing_yards)  # Prints Brees' career passing yards
# Prints Tom Brady's passing yards total only for the 2017 season
print(tombrady('2017').passing_yards)
# Prints a Pandas DataFrame of all relevant stats per season for Brees
print(tombrady.dataframe)
````

The `Schedule` module can be used to iterate over all games in a team’s schedule to get game information such as the date, score, result, and more. Each game also has a link to the Boxscore class which has much more detailed information on the game metrics.

````python
from sportsreference.nfl.schedule import Schedule

houston_schedule = Schedule('NWE')
for game in houston_schedule:
    print(game.date)  # Prints the date the game was played
    print(game.result)  # Prints whether the team won or lost
    # Creates an instance of the Boxscore class for the game.
    boxscore = game.boxscore
````

## NBA

This cell combs through every NBA team and finds and displays the tallest player on each squad.

````python
from sportsreference.nba.teams import Teams

def get_height_in_inches(height):
    feet, inches = height.split('-')
    return int(feet) * 12 + int(inches)

def print_tallest_player(team_heights):
    tallest_player = max(team_heights, key=team_heights.get)
    print('%s: %s in.' % (tallest_player, team_heights[tallest_player]))

for team in Teams():
    print('=' * 80)
    print(team.name)
    team_heights = {}
    for player in team.roster.players:
        height = get_height_in_inches(player.height)
        team_heights[player.name] = height
    print_tallest_player(team_heights)
````

## MLB

The cell below iterates through every game in the Boston Red Sox schedule for the most recent season and both prints the boxscore index (like the BOS/BOS20180808020 example we saw above) as well as creates a boxscore object representing the specific game in the iteration which can be used in the same way as the previous example

````python
from sportsreference.mlb.schedule import Schedule

boston_schedule = Schedule('BOS')
for game in boston_schedule:
    print(game.boxscore_index)  # Prints the boxscore URI for each game
    boxscore = game.boxscore  # Returns an instance of the Boxscore class for this specific game
````

You can also retrieve the URI of BOS/BOS20180808020 by navigating to sports-reference.com and select a game played on August 2, 2018 between the Boston Red Sox and the New York Yankees. 

The URL for this specific game is [this](https://www.baseball-reference.com/boxes/BOS/BOS201808020.shtml).

````python
from sportsreference.mlb.boxscore import Boxscore

game_data = Boxscore('BOS/BOS201808020')
print(game_data.away_runs, game_data.home_runs)
print(game_data.dataframe)
````

Here, we iterate from the 2000 season until the 2019 season, and check the number of wins each MLB team gained each year and print the team with the most victories.

````python
from sportsreference.mlb.teams import Teams

def print_most_wins(year, wins):
    most_wins = max(wins, key=wins.get)
    print('%s: %s - %s' % (year, wins[most_wins], most_wins))

for year in range(2000, 2019):
    wins = {}
    for team in Teams(year):
        wins[team.name] = team.wins
    print_most_wins(year, wins)
````
