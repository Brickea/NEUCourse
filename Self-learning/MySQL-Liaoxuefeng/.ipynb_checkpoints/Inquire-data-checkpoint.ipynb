## Condition Inquire

条件	表达式举例1	表达式举例2	说明  
使用=判断相等	score = 80	name = 'abc'	字符串需要用单引号括起来  
使用>判断大于	score > 80	name > 'abc'	字符串比较根据ASCII码，中文字符比较根据数据库设置  
使用>=判断大于或相等	score >= 80	name >= 'abc'	  
使用<判断小于	score < 80	name <= 'abc'	  
使用<=判断小于或相等	score <= 80	name <= 'abc'	  
使用<>判断不相等	score <> 80	name <> 'abc'	  
使用LIKE判断相似	name LIKE 'ab%'	name LIKE '%bc%'	%表示任意字符，例如'ab%'将匹配'ab'，'abc'，'abcd'  

```python
%%sql
SELECT * FROM students WHERE score>=90;
```

```python
%%sql
SELECT * FROM students WHERE score >= 80 AND gender = 'M';
```

```python
%%sql
SELECT * FROM students WHERE score >=80 or gender='M';
```

```python
%%sql
SELECT * FROM students WHERE NOT class_id = 2;
```

```python
%%sql
SELECT * FROM students WHERE (score < 80 OR score > 90) AND gender = 'M';
```

```python
%%sql
SELECT * FROM students WHERE gender <> 'M';
```

## Paging Inquire

```python
%%sql
# Limit will only show the first several rows you choose
SELECT *
FROM students
LIMIT 2;
```

```python
%%sql
# OFFSET will get the result after the index you choose
SELECT *
FROM students
LIMIT 2
OFFSET 2;
```

```python
%%sql
# If we choose to use 3 rows as our page display row
# We use count to calculate the pagesize
SELECT CEILING(COUNT(*)/3)
FROM students;
```

## Aggregate inquire

SUM	计算某一列的合计值，该列必须为数值类型  
AVG	计算某一列的平均值，该列必须为数值类型  
MAX	计算某一列的最大值  
MIN	计算某一列的最小值  
注意，MAX()和MIN()函数并不限于数值类型。如果是字符类型，MAX()和MIN()会返回排序最后和排序最前的字符。

```python
%%sql
SELECT COUNT(*) FROM students;
```

```python
%%sql
SELECT COUNT(*) num
FROM students;
```

```python
%%sql
SELECT COUNT(*) boys
FROM students
WHERE gender = 'M';
```

```python
%%sql
SELECT AVG(score) boys_average_score
FROM students
WHERE gender='M';
```

## Group

```python
%%sql
SELECT COUNT(*) num
FROM students
GROUP BY class_id;
```

```python
%%sql
SELECT class_id, COUNT(*) num
FROM students
GROUP BY class_id
```

```python
%%sql
SELECT name, class_id, COUNT(*) num FROM students GROUP BY class_id;
```

Not surprisingly, we will get a syntax error when executing this query, because in any grouping, only the class_id is the same, and the name is different. The SQL engine cannot put the values of multiple names into a row of records. Therefore, only the grouped columns can be placed in the columns of the aggregate query.

```python
%%sql
SELECT class_id, gender, COUNT(*) num
FROM students
GROUP BY gender,class_id;
```

```python
%%sql
# Set alias
SELECT
    students.id sid,
    students.name,
    students.gender,
    students.score,
    classes.id cid,
    classes.name cname
FROM students, classes;
```

```python
%%sql
# It can be more concise
SELECT
    s.id sid,
    s.name,
    s.gender,
    s.score,
    c.id cid,
    c.name cname
FROM students s, classes c;
```

```python
%%sql
SELECT
    s.id sid,
    s.name,
    s.gender,
    s.score,
    c.id cid,
    c.name cname
FROM students s, classes c
WHERE s.gender = 'M' AND c.id = 1;
```

## Join Inquire

### INNER JOIN

注意INNER JOIN查询的写法是：

先确定主表，仍然使用FROM <表1>的语法;  
再确定需要连接的表，使用INNER JOIN <表2>的语法； 
然后确定连接条件，使用ON <条件...>，这里的条件是s.class_id = c.id，表示students表的class_id列与classes表的id列相同的行需要连接；  
可选：加上WHERE子句、ORDER BY等子句。  
使用别名不是必须的，但可以更好地简化查询语句。  

```python
%%sql
# Get all students information with class name
SELECT s.id, s.name, s.class_id, c.name class_name, s.gender, s.score
FROM students s
INNER JOIN classes c
ON s.class_id = c.id;
```

### OUTER JOIN

```python
%%sql
SELECT s.id, s.name, s.class_id, c.name class_name, s.gender, s.score
FROM students s
FULL OUTER JOIN classes c
ON s.class_id = c.id;
```
